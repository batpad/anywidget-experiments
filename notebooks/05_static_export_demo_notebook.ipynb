## Linked CounterWidgets

A source counter linked to a follower. The follower should mirror the source's value changes, all client-side.

```python
source = CounterWidget(label="Source Counter", widget_id="static_source", value=0)
source
```

```python
follower = LinkedCounterWidget(
    label="Follower Counter",
    widget_id="static_follower",
    link_to="static_source",
    link_mode="mirror",
    value=0
)
follower
```

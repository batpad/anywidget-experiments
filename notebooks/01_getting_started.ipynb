# Getting Started with Anywidget

This notebook demonstrates the basics of anywidget and inter-widget communication patterns.

## Setup

First, let's import the necessary libraries and set up our environment.

```python
import sys
import pathlib

# Add parent directory to path to import our widgets
sys.path.insert(0, str(pathlib.Path().absolute().parent))

from widgets.counter_widget import CounterWidget
from widgets.linked_counter import LinkedCounterWidget
import ipywidgets as widgets
from IPython.display import display
```

## Simple Counter Widget

Let's start with a simple counter widget that demonstrates the basic anywidget pattern.

```python
# Create a simple counter widget
counter1 = CounterWidget(
    value=0,
    label="My First Counter",
    widget_id="counter_1"
)

counter1
```

### Interacting with the Widget from Python

We can interact with the widget from Python code:

```python
# Get current value
print(f"Current value: {counter1.value}")

# Increment the counter
counter1.increment(5)
print(f"After increment(5): {counter1.value}")

# Decrement the counter
counter1.decrement(2)
print(f"After decrement(2): {counter1.value}")

# Reset the counter
# counter1.reset()
print(f"After reset: {counter1.value}")
```

## Multiple Widgets

Let's create multiple counter widgets to see how they can work independently:

```python
# Create multiple counters
counter2 = CounterWidget(
    value=10,
    label="Counter 2",
    widget_id="counter_2"
)

counter3 = CounterWidget(
    value=20,
    label="Counter 3",
    widget_id="counter_3"
)

# Display them in a horizontal layout
widgets.HBox([counter1, counter2, counter3])
```

## Inter-Widget Communication

Now let's demonstrate inter-widget communication using our LinkedCounterWidget:

```python
# Create a linked counter that mirrors counter_2
linked1 = LinkedCounterWidget(
    value=0,
    label="Linked to Counter 2",
    widget_id="linked_1",
    link_to="counter_2",
    link_mode="mirror"
)

# Create another linked counter that shows the sum
linked2 = LinkedCounterWidget(
    value=5,
    label="Sum with Counter 3",
    widget_id="linked_2",
    link_to="counter_3",
    link_mode="sum"
)

# Display the original counter and linked widgets
display(widgets.VBox([
    widgets.HTML("<h3>Original Counters</h3>"),
    widgets.HBox([counter2, counter3]),
    widgets.HTML("<h3>Linked Counters</h3>"),
    widgets.HBox([linked1, linked2])
]))
```

## Observing Widget Changes

We can observe changes to widget values from Python:

```python
# Create a widget to display change events
output = widgets.Output()

def on_value_change(change):
    with output:
        print(f"Value changed from {change['old']} to {change['new']}")

# Observe changes on counter1
counter1.observe(on_value_change, names='value')

display(widgets.VBox([counter1, output]))
```

## Chain of Linked Widgets

Let's create a chain of widgets where each one depends on the previous:

```python
# Create a chain of linked widgets
chain_source = CounterWidget(
    value=1,
    label="Source",
    widget_id="chain_source"
)

chain_mirror = LinkedCounterWidget(
    value=0,
    label="Mirror of Source",
    widget_id="chain_mirror",
    link_to="chain_source",
    link_mode="mirror"
)

chain_sum = LinkedCounterWidget(
    value=10,
    label="Sum with Mirror",
    widget_id="chain_sum",
    link_to="chain_mirror",
    link_mode="sum"
)

chain_diff = LinkedCounterWidget(
    value=100,
    label="Difference from Sum",
    widget_id="chain_diff",
    link_to="chain_sum",
    link_mode="diff"
)

# Display the chain
widgets.VBox([
    widgets.HTML("<h3>Widget Chain Demo</h3>"),
    widgets.HTML("<p>Change the source value and watch it propagate!</p>"),
    chain_source,
    widgets.HTML("↓"),
    chain_mirror,
    widgets.HTML("↓"),
    chain_sum,
    widgets.HTML("↓"),
    chain_diff
])
```

## Widget State Persistence

The widget state is preserved when the notebook is saved. This is crucial for static HTML exports:

```python
# Create a widget with a specific state
persistent = CounterWidget(
    value=42,
    label="Persistent Counter",
    widget_id="persistent_counter"
)

persistent
```

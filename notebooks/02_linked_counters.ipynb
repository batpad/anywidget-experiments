# 2. Anywidgets talking to each other

Two widgets on the same page can communicate without a kernel by going through a shared in-page registry. Our convention: each widget's JS `render` registers itself in `window.__widgetRegistry` keyed by its `widget_id` trait, and listens to a `window.__widgetEvents` `EventTarget` for cross-widget events. A separate `window.__myst_widgets` registry (populated by the static-export plugin) lets widgets find each other by `widget_id`, `_anywidget_id`, or `model_id`.

Below, a `LinkedCounterWidget` watches a `CounterWidget` and mirrors its value. Click the source's `+`/`-` and the follower's value follows.

```python
import sys
import pathlib
sys.path.insert(0, str(pathlib.Path().absolute().parent))

from widgets.counter_widget import CounterWidget
from widgets.linked_counter import LinkedCounterWidget
```

## Source

Same `CounterWidget` as before. Its `widget_id` is what the follower will look up.

```python
source = CounterWidget(
    label="Source",
    widget_id="linked_source",
    value=0,
)
source
```

## Follower

`LinkedCounterWidget` looks up its `link_to` target in `window.__widgetRegistry` and subscribes to its `value` changes. `link_mode="mirror"` copies the source's value verbatim. (Other modes: `sum`, `diff`.)

```python
follower = LinkedCounterWidget(
    label="Follower",
    widget_id="linked_follower",
    link_to="linked_source",
    link_mode="mirror",
    value=0,
)
follower
```
